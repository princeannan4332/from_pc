```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

df=pd.read_excel('NewChanges.xlsx')
print(df)
```

# Distribution plot

```python
#distribution plot counts the number of a range eg:3-4
#KDE false removes the line drawn over the graph
#bins just divides the information well to make readabity easy
#color cahnges the colour of the bars
sns.distplot(df['Rating'],kde=False,color='green',bins=20)
```

# Joint Plot

```python
#kind=reg produces a regression line 

sns.jointplot(x='Speed',y='Power',data=df,kind='reg')
```

# KDE PLOT

```python
sns.kdeplot(df['Power'])
```

# Pair Plot

```python
sns.pairplot(df,hue='Skills')
```

# RUG PLOT

```python
sns.rugplot(df['Speed'],color='red')
```

# Styling

```python
#.set_style changes the background of the graph properties:darkgrid,white,ticks
sns.set_style('white')

#figsize specifies [width,height] of graph
plt.figure(figsize=[20,4])

#.set_context can be set to:paper,talk,poster
#.set_context font_scale specifies the size of the font
sns.set_context('paper',font_scale=1.5)
sns.jointplot(x='Resident',y='Speed',data=df,color='red')

#removing the borders of the x or y axis
sns.despine(left=False,right=False)
```

```python
df['count']=0
j=df.groupby('Resident').count()['count']
print(j)
```

# Bar Plot

```python
sns.set_context('paper')
sns.barplot(x='Speed',y='Skills',data=df)
```
